```python
# Manejar valores nulos (NaN):

# Estrategia 1: borrarlos. Si el dataset tiene pocas filas y hay muchos nulos, borrarlos igual no es la mejor opción
# df = df.dropna()

# Estrategia 2: imputarlos o rellenarlos con una estadística global de toda la columna
# numérica: media, mediana, predicción con KNNImputer
# df['gasto_total'].fillna(df['gasto_total'].mean())
# median() por defecto ya excluye los nulos
df['gasto_total'].fillna(df['gasto_total'].median()) # más robusto

# Estrategia 3: imputarlos o rellenarlos con una estadística de esa columna pero calculada filtrando por filas más similares
# df.apply() combinado con una función que use if para detectar nulos y filtrar por filas similares para calcular la estadísticas

# Estrategia 4: utilizar algoritmos para predecir los valores faltantes: KNNImputer, IterativeImputer
```

```python
df = df.fillna({
    'gasto_total': df['gasto_total'].median(),
    'size': round(df['size'].median(), 0)
})
df['size'] = df['size'].astype(np.int8)
```

* numérico: media, mediana, predicción (regresión)
* categórico: moda, valor centinela ('Otros'), predicción (clasificación)
* fecha: ffill o bfill

Sobre todo nos interesa eliminar o imputar valores nulos de cara a la fase de modelado, porque cuando queremos entrenar un modelo Scikit Learn si el dataset tiene nulos Scikit no deja entrenar te da un error. 

* Borrar columnas que no nos sirvan o que sean en su mayoría valores nulos
* Borrar filas que tengan nulos solo si tenemos muchos datos y el borrado no te afecta 
* Rellenar nulos de filas si tenemos pocos datos y borrarlos afectaría a nivel global

Programación defensiva, robusta. 

En Scikit usaremos el concepto de Pipelines donde se hace la imputación de nulos y preprocesados y el modelado todo junto en un único pipeline o sucesión de pasos que generan un único objeto con el modelo para hacer predicciones.

```python
# encoding: convertir categóricos a numéricos
# ventaja: no introduce nuevas columnas
# ideal: modelos basados en árboles: CART, random forest, XGBoost 
encoding = {
    'Thur': 1,
    'Fri': 2,
    'Sat': 3,
    'Sun': 4
}
df['day_int'] = df['day'].map(encoding)
df['day_int'].value_counts()
```
