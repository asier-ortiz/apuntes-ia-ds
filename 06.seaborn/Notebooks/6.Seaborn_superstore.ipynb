```python
numeric_columns = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost']

for column in numeric_columns:
    plt.figure(figsize=(6,4))
    sns.histplot(data=df, x=column)
    plt.title(f'Hisplot de {column}')

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    plt.xlim(lower_limit, upper_limit)

    plt.show()
```

## Análisis bivariante

### barplot: Sales / Profit totales por año

2011-2015

```python
# Crear columna Year
df = pd.read_csv('../../../data/superstore_dataset2011-2015.csv')
# Pandas 2.0.1
# df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed')
# Pandas 1.5
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['year'] = df['Order Date'].dt.year

# groupby de Sales por Year
sales_per_year = df.groupby('year')['Sales'].sum() # serie
profit_per_year = df.groupby('year')['Profit'].sum() # serie

# barplot
sns.barplot(x=sales_per_year.index, y=sales_per_year.values)
```

### barplot de sales_per_year y profit_per_year

```python
# Crear un DataFrame de sales_per_year y mostrar una gráfica de área
df_sales_per_year = pd.DataFrame({
    'Sales': sales_per_year.values, # serie
    'Profit': profit_per_year.values # serie
}, index=sales_per_year.index) # índice

df_sales_per_year.plot(kind='bar', alpha=0.5)
plt.title('Ventas y Beneficios por año')
plt.ticklabel_format(style='plain', axis='y')  # Desactivar la notación científica en el eje Y
```

```python
# Crear un DataFrame de sales_per_year y mostrar una gráfica de área
df_sales_per_year = pd.DataFrame({
    'Sales': sales_per_year.values, # serie
    'Profit': profit_per_year.values # serie
}, index=sales_per_year.index) # índice

df_sales_per_year.plot(kind='line', alpha=0.5)
plt.title('Ventas y Beneficios por año')
plt.ticklabel_format(style='plain', axis='y')  # Desactivar la notación científica en el eje Y
```

```python
# seaborn
fig, ax1 = plt.subplots(figsize=(10, 6))

# barplot para sales_per_year
ax1.set_ylabel('Sales')
sns.barplot(x=profit_per_year.index, y=profit_per_year.values, ax=ax1, alpha=0.5, color='blue')
plt.ticklabel_format(style='plain', axis='y')

# barplot para profit_per_year
ax2 = ax1.twinx() # eje X compartido
sns.barplot(x=sales_per_year.index, y=sales_per_year.values, ax=ax2, alpha=0.5, color='yellow')
ax2.set_ylabel('Profit');
```

```python
fig, ax1 = plt.subplots()

## sales

color = 'blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('Sales', color=color)
ax1.bar(df_sales_per_year.index, df_sales_per_year['Sales'], color=color) # Barras
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'red'
ax2.set_ylabel('Profit', color=color)
ax2.plot(df_sales_per_year.index, df_sales_per_year['Profit'], color=color) # Líneas
ax2.tick_params(axis='y', labelcolor=color)
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

# lineplot para sales_per_year
sns.lineplot(x=sales_per_year.index, y=sales_per_year.values, ax=ax1, alpha=0.5, marker='o')
ax1.set_ylabel('Sales')
plt.ticklabel_format(style='plain', axis='y')

# lineplot para profit_per_year
ax2 = ax1.twinx() # eje X compartido
sns.lineplot(x=profit_per_year.index, y=profit_per_year.values, ax=ax2, alpha=0.5, marker='o', color='orange')
ax2.set_ylabel('Profit')

plt.show()
```
