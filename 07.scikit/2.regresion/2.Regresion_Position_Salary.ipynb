```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('../../data/salaries_position.csv')
df.head()
```

```python
sns.scatterplot(data=df, x='Level', y='Salary')
```

```python
X = df[['Level']] # dataframe
y = df['Salary'] # serie 

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
sns.scatterplot(data=df, x='Level', y=df['Salary'])
sns.lineplot(x=df['Level'], y=y_pred, color='r', label='Predictions')
```

```python
model.predict([[10]])
```

```python
from sklearn.preprocessing import PolynomialFeatures

# Transformar la entrada X agregando nuevas caracter√≠sticas que son el cuadrado de las variables de entrada originales
preprocessor = PolynomialFeatures(degree=3)
X_g2 = preprocessor.fit_transform(X)
X_g2
```

```python
model = LinearRegression()
model.fit(X_g2, y)

y_pred = model.predict(X_g2)
```

```python
sns.scatterplot(data=df, x='Level', y=df['Salary'])
sns.lineplot(x=df['Level'], y=y_pred, color='r', label='Predictions')
```
