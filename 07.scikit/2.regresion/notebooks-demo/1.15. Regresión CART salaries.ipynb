# Árboles de decisión (CART)

Ejemplo de regresión con DecisionTreeRegressor

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
dataset = pd.read_csv('../data/salaries_position.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
print(X)
```

```python
print(y)
```

```python
plt.scatter(X, y)
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
# regressor = DecisionTreeRegressor(random_state=0, max_depth=2)
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X,y)
```

```python
y_pred = regressor.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, c='r')

plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
regressor.predict(np.array([[2]]))
```

```python
r2 = r2_score(y, y_pred)
r2
```

```python
mean_absolute_error(y, y_pred)
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Check It (Regression Model)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
X_grid
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Example of Decision Regression Model')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn import tree
text_representation = tree.export_text(regressor)
print(text_representation)
```

```python
with open("decision_tree.log", "w") as fout:
    fout.write(text_representation)
```
